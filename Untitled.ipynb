```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('neos.csv')

plt.figure(figsize=(17,17))

sns.heatmap(data.corr(),
            linewidths = 0.25,
            square = False,
            cmap = 'inferno',
            linecolor = 'black',
            annot= True);

plt.show()
```

```python
plt.scatter(data['estimated_max_diameter_kilometers'], data['miss_distance_kilometers'], alpha = 1)
plt.xlabel("estimated_max_diameter_kilometers")
plt.ylabel("miss_distance_kilometers");

plt.show()
```
