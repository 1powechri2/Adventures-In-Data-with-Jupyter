```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('neos.csv')
```

```python
data.head(5)
```

```python
plt.figure(figsize=(17,17))

sns.heatmap(data.corr(),
            linewidths = 0.25,
            square = False,
            cmap = 'inferno',
            linecolor = 'black',
            annot= True);
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data['estimated_max_diameter_kilometers'], data['miss_distance_kilometers'], alpha = 1, color = 'red');
plt.xlabel("estimated_max_diameter_kilometers", fontsize=20);
plt.ylabel("miss_distance_kilometers", fontsize=20);
plt.title("Asteroid Size vs Proximity to Earth", fontsize=30);
plt.grid();

plt.show()
```

```python
small_neos = data['estimated_max_diameter_kilometers'] < 1
actual_small_neos = data['estimated_max_diameter_kilometers'][small_neos]
small_neos_miss_dist = data[actual_small_neos]['miss_distance_kilometers']

actual_small_neos.size
```

```python
small_neos_miss_dist.size
```

```python
plt.figure(figsize=(20,10))
plt.scatter(actual_small_neos, small_neos_miss_dist, alpha = 1, color = 'red');
plt.xlabel("estimated_max_diameter_kilometers", fontsize=20);
plt.ylabel("miss_distance_kilometers", fontsize=20);
plt.title("Asteroid Size vs Proximity to Earth", fontsize=30);
plt.grid();

plt.show()
```
