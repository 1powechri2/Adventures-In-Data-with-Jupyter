```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('neos.csv')
```

```python
data.head(5)
data.shape
```

```python
plt.figure(figsize=(15,17))

sns.heatmap(data.corr(),
            linewidths = 0.25,
            square = False,
            cmap = 'inferno',
            linecolor = 'black',
            annot= True);
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data['estimated_max_diameter_kilometers'], data['miss_distance_kilometers'], alpha = 1, color = 'red');
plt.xlabel("estimated_max_diameter_kilometers", fontsize=20);
plt.ylabel("miss_distance_kilometers", fontsize=20);
plt.title("Asteroid Size vs Proximity to Earth", fontsize=30);
plt.grid();

plt.show()
```

```python
small_neos = data['estimated_max_diameter_kilometers'] < 1
actual_small_neos = data['estimated_max_diameter_kilometers'][small_neos]
small_neos_miss_dist = data['miss_distance_kilometers'][small_neos]

actual_small_neos.size
```

```python
small_neos_miss_dist.size
```

```python
plt.figure(figsize=(20,10))
plt.scatter(actual_small_neos, small_neos_miss_dist, alpha = 1, color = 'red');
plt.xlabel("estimated_max_diameter_kilometers", fontsize=20);
plt.ylabel("miss_distance_kilometers", fontsize=20);
plt.title("Asteroid Size vs Proximity to Earth", fontsize=30);
plt.grid();

plt.show()
```

```python
plt.figure(figsize=(20,10))
sorted_small_neos = sorted(actual_small_neos)
plt.plot(sorted_small_neos)
plt.grid()
plt.show()
```

```python
data['close_approach_date'] = pd.to_datetime(data['close_approach_date'])

mask = (data['close_approach_date'] >= pd.to_datetime('2018-1-01')) & (data['close_approach_date'] <= pd.to_datetime('2018-1-31'))
        
jan_2018_neos = data.loc[mask]

split_hazard_neos = jan_2018_neos.groupby('is_potentially_hazardous')

jan_2018_hazardous_neos = split_hazard_neos.get_group(True)

jan_2018_hazardous_neos.shape
```

```python
import calendar

plt.figure(figsize=(20,10))

diameter_x_100 = jan_2018_hazardous_neos['estimated_max_diameter_kilometers'] * 100

plt.scatter(jan_2018_hazardous_neos['close_approach_date'], jan_2018_hazardous_neos['estimated_max_diameter_kilometers'], 
s = diameter_x_100, c = diameter_x_100, alpha = 1);

plt.xticks(jan_2018_hazardous_neos['close_approach_date'], jan_2018_hazardous_neos['close_approach_date'],
rotation=90)

plt.ylabel("miss_distance_kilometers", fontsize=20);
plt.title("Hazardous Asteroids January 2018", fontsize=30);
plt.grid();

plt.show()
```
